Solucionando o problema do operador XOR (Hello Word do Deep Learning antes do MNIST)

Importação das Bibliotecas

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
```

Normalização dos vetores de entrada e saída do modelo com a biblioteca numpy

```python
train_x = np.array([[0,0],[0,1],[1,0],[1,1]], ndmin=2)
train_y = np.array([[0],[1],[1],[0]])
```

# Definindo a arquitetura do modelo
RNA PMC com arquitetura de 2 camadas processadoras:


20 neurônios na camada intermediaria


1 neurônio na camada de saída

```python
model = Sequential()
model.add(Dense(20, input_dim=len(train_x[0]), activation='relu'))
model.add(Dense(len(train_y[0]), activation='relu'))
```

Compilando o modelo

```python
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
model.fit(train_x, train_y, batch_size=4, epochs=200)
```

Realizando a predição

```python
teste = np.array([[0,0],[0,1],[1,0],[1,1]], ndmin=2)
```

```python
model.predict(teste, batch_size=1, verbose=0)
```
